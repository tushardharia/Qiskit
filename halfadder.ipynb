```python
from qiskit import *
from qiskit.visualization import plot_histogram

half_adder = QuantumCircuit(4,2)

half_adder.x(0)
half_adder.x(1)

half_adder.barrier()
half_adder.cx(0,2)
half_adder.cx(1,2)
half_adder.ccx(0,1,3)
half_adder.barrier()

half_adder.measure(3,1)
half_adder.measure(2,0)

half_adder.draw(output="mpl")
```

```python
counts = execute(half_adder, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
